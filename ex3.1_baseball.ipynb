```python
def plot_hist_each_column(df):
    plt.rcParams['figure.figsize'] = [20, 16]
    fig = plt.figure(1)
    # df의 열 개수 만큼의 subplot을 출력
    for i in range(len(df)):
        ax = fig.add_subplot(5, 5, i+1)
        plt.hist(df[df.columns[i]], bins=50)
        ax.set_title(df.columns[i])
    plt.show()
```

```python
plot_hist_each_column(picher_features)
```

```python
# 판다스 형태로 정의된 데이터를 출력할 때 scientific-notation이 아닌 float 모양으로 출력되게 해줌
pd.options.mode.chained_assignment = None

# 직접 정규화
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x : (x-series_mean) / series_std)
    return df

scale_columns = picher_features.columns.difference(['연봉(2018)'])
scale_columns
```

```python
picher_scaling = standard_scaling(picher, scale_columns)
picher_scaling
```

```python
picher_scaling.info()
```

```python
# 팀명 피처를 원-핫 인코딩
team_encoding = pd.get_dummies(picher['팀명'])
picher_scaling = picher.drop('팀명', axis = 1)
picher_scaling = picher_scaling.join(team_encoding)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

X = picher_scaling[picher_scaling.columns.difference(['선수명', '연봉(2018)'])]    # 제외하고 
y = picher_scaling['연봉(2018)']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)   # train_size : test_size 비율 = 8:2
```

```python
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
```

```python
lr.coef_    # .reshape(-1,4) 행의 개수 모를 때 : -1
```

```python
import statsmodels.api as sm

# statsmodel 라이브러리 
X_train = sm.add_constant(X_train)
# R-squared
model = sm.OLS(y_train, X_train).fit()
model.summary()
```

```python
# 한글 출력
mpl.rc('font', family = 'NanumGothic')
plt.rcParams['figure.figsize'] = [20, 16]

# 회귀 계수 리스트
coefs = model.params.tolist()
coefs_series = pd.Series(coefs)

# 변수명 리스트
x_labels = model.params.index.tolist()

# 회귀 계수 출력
ax = coefs_series.plot(kind = 'bar')    # 막대 그래프
ax.set_title('feature_coef_graph')
ax.set_xlabel('x_features')
ax.set_ylabel('coefs')
ax.set_xticklabels(x_labels)
```

```python
X = picher_scaling[picher_scaling.columns.difference(['선수명', '연봉(2018)'])]
y = picher_scaling['연봉(2018)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 19)

lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
```

```python
# R2 score
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
# RMSE score
y_train_predictions = lr.predict(X_train)
y_test_predictions = lr.predict(X_test)

print(sqrt(mean_squared_error(y_train, y_train_predictions)))
print(sqrt(mean_squared_error(y_test, y_test_predictions)))
```

```python
# 원래의 데이터 프레임에 2018년 연봉 정보를 merge
result = picher_scaling.sort_values(by=['연봉(2018)'], ascending = False)
result.drop(['연봉(2017)'], axis = 1, inplace = True, errors = 'ignore')
result = result.merge(picher, on = ['선수명'], how = 'left')
result
```

```python
result_df = result[['선수명', '연봉(2018)', '예측연봉(2018)', '연봉(2017)']]
result_df.columns = ['선수명', '실제연봉(2018)', '예측연봉(2018)', '작년연봉(2017)']
result_df
```

```python
# 재계약하여 연봉이 변화한 선수만을 대상으로 관찰
result_df = result_df[result_df['작년연봉(2017)'] != result_df['실제연봉(2018)']]
result_df = result_df.iloc[:10, :]
result_df.head()
```
